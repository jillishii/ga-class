```python
# function to clean first shop in the shops list
def remove_first_two(a_string):
    new_string = a_string[2:]
    return new_string
```

```python
# function to clean last shop in the shops list
def remove_last_two(a_string):
    new_string = a_string[:-2]
    return new_string
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
import seaborn as sns
```

```python
occurances = category_df['top_category'].value_counts()
```

```python
objects = []
for i in range(15):
    objects.append(occurances.index[i])

y_pos = np.arange(len(objects))
plt.bar(y_pos, occurances, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of Subcategories')
plt.xlabel('Top Category')
plt.title('Number of Subcategories in each Top Category')
plt.show()
```

The second data file contains a list of shops and their corresponding data for items reviews.

```python
# import csv
shop_df = pd.read_csv("shop_data.csv")
```

```python
shop_df.drop(shop_df.columns[[0, 1]], axis=1)
```

```python
shop_df['review_dates'][0]
```

```python
# takes raw list containing shop names
#returns clean list with shop name only

import re

def clean_name(messy_name):
    new_list = []
    #for name in messy_name_list:
    finder = re.compile("', '" + "(.*?)" + "', '", re.IGNORECASE)
    matches = finder.findall(messy_name)
    no_replace = re.compile("\s+")
    p = no_replace.sub("", matches[0])
    new_list.append(p)
    return new_list
```

```python
clean_name(shop_df['review_dates'])
```
